1. Averaging

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image
img = cv2.imread('bottle.jpg')

# Apply blur
blurred_img = cv2.blur(img, (5, 5))

# Show images
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Blurred')
plt.axis('off')

plt.show()
```

2. Gaussian Blurring

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('bottle.jpg')
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
gaussian_blur = cv2.GaussianBlur(src=img, ksize=(3,3),sigmaX=0, sigmaY=0)
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB))
plt.title('Blurred')
plt.axis('off')
plt.show()
```

3. Median Blurring

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
# Read the image
img = cv2.imread('bottle.jpg')
# Apply Median blur
blurred_img = cv2.medianBlur(img, 5)
# Show images
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Blurred')
plt.axis('off')
plt.show()
```

4.bilateral filtering

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image
img = cv2.imread('bottle.jpg')

# Apply bilateral filter
blurred_img = cv2.bilateralFilter(img, 9, 75, 75)

# Show images
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Blurred')
plt.axis('off')
plt.show()
```

5.Sobel filtering

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image
img = cv2.imread('bottle.jpg')

# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Sobel filter for edge detection
sobel_x = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=3)

# Calculate the magnitude of the gradients
sobel_magnitude = cv2.magnitude(sobel_x, sobel_y)

# Show images
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Sobel Filtered')
plt.axis('off')
plt.show()
```

6.Emboss filtering

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('bottle.jpg')
plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.axis('off')

emboss_kernel = np.array([[-1, 0, 0],
                    [0, 0, 0],
                    [0, 0, 1]])

emboss_img = cv2.filter2D(src=img, ddepth=-1, kernel=emboss_kernel)
plt.subplot(1, 2, 2)
plt.title('Blurred')
plt.imshow(cv2.cvtColor(emboss_img, cv2.COLOR_BGR2RGB))
# plt.imshow(emboss_img)
plt.axis('off')
plt.show()
```
